```python
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 0) Make sure weâ€™re in your project
os.chdir(os.path.expanduser("~/Desktop/androgen_docking_project"))

# 1) Load the Excel you exported
path = Path("3_gene_expression/GSE216372_all_genes_all_samples_ARKO_vs_WT.xlsx")  # or GSE216372_fixed.xlsx if you used that name
df = pd.read_excel(path)

# 2) Sanity check: key columns present?
needed = {"GeneID","Genename","log2FoldChange","padj"}
missing = [c for c in needed if c not in df.columns]
print("Missing columns:", missing)  # should be []

# 3) Clean up gene name column (fallback to GeneID if name is missing)
gene = df["Genename"].astype(str)
gene = np.where(gene.isna() | (gene=="nan") | (gene==""), df["GeneID"].astype(str), gene)
df["_GENE"] = gene

# 4) Basic filters and ranking
# Convention: positive log2FC = higher in ARKO; negative log2FC = higher in WT
sig = df.dropna(subset=["log2FoldChange","padj"]).copy()
sig = sig.sort_values("log2FoldChange", ascending=False)

# 5) Save a ranked CSV for your records
outdir = Path("4_enrichment_outputs"); outdir.mkdir(exist_ok=True)
sig[["_GENE","Genename","GeneID","log2FoldChange","padj","baseMean"]].to_csv(outdir/"GSE216372_ranked_all.csv", index=False)

# 6) Make simple UP/DOWN lists for enrichment (padj < 0.05; top 150 by effect size)
alpha = 0.05
up = sig[(sig["padj"] < alpha) & (sig["log2FoldChange"] > 0)].copy()
down = sig[(sig["padj"] < alpha) & (sig["log2FoldChange"] < 0)].copy()

up_top = up.nlargest(150, "log2FoldChange")
down_top = down.nsmallest(150, "log2FoldChange")  # most negative

# Save gene lists (one gene per line) â€” Enrichr/GSEA-ready
(up_top["_GENE"]).to_csv(outdir/"top150_up_genes.txt", index=False, header=False)
(down_top["_GENE"]).to_csv(outdir/"top150_down_genes.txt", index=False, header=False)

print(f"Saved:\n- {outdir/'GSE216372_ranked_all.csv'}\n- {outdir/'top150_up_genes.txt'}\n- {outdir/'top150_down_genes.txt'}")

# 7) Quick plots (save to 5_figures)
figdir = Path("5_figures"); figdir.mkdir(exist_ok=True)

# Top 10 up/down barplots
top10_up = up.nlargest(10, "log2FoldChange")
top10_down = down.nsmallest(10, "log2FoldChange")

plt.figure()
plt.bar(top10_up["_GENE"], top10_up["log2FoldChange"])
plt.xticks(rotation=90)
plt.ylabel("log2FC (ARKO vs WT)")
plt.title("Top 10 Upregulated (ARKO > WT)")
plt.tight_layout()
plt.savefig(figdir/"GSE216372_top10_up.png", dpi=300)
plt.show()

plt.figure()
plt.bar(top10_down["_GENE"], top10_down["log2FoldChange"])
plt.xticks(rotation=90)
plt.ylabel("log2FC (ARKO vs WT)")
plt.title("Top 10 Downregulated (ARKO < WT)")
plt.tight_layout()
plt.savefig(figdir/"GSE216372_top10_down.png", dpi=300)
plt.show()

# 8) (Optional) Volcano plot for a quick overview
# Avoid log(0) by clipping p-values
p = sig["padj"].clip(lower=1e-300)
plt.figure()
plt.scatter(sig["log2FoldChange"], -np.log10(p), s=6, alpha=0.6)
plt.axhline(-np.log10(alpha), linestyle="--", linewidth=1)
plt.axvline(0, linestyle="--", linewidth=1)
plt.xlabel("log2 Fold Change (ARKO vs WT)")
plt.ylabel("-log10(adj p-value)")
plt.title("Volcano Plot â€” GSE216372")
plt.tight_layout()
plt.savefig(figdir/"GSE216372_volcano.png", dpi=300)
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

os.chdir(os.path.expanduser("~/Desktop/androgen_docking_project"))

FIG_W, FIG_H = 15, 7.2
TITLE_FS_MAIN = 18
TITLE_FS_RIGHT = 17
BODY_FS  = 13

plt.figure(figsize=(FIG_W, FIG_H))
ax = plt.gca()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

# --- Boxes: give them more horizontal spacing ---
box_lig  = (0.04, 0.56, 0.27, 0.30)
box_bind = (0.37, 0.56, 0.27, 0.30)
box_expr = (0.73, 0.42, 0.23, 0.44)   # moved right, a bit narrower

def draw_box(b):
    x,y,w,h = b
    ax.add_patch(mpatches.FancyBboxPatch(
        (x,y), w,h, boxstyle="round,pad=0.02",
        fc="#f3f4f6", ec="gray", lw=1.6
    ))

for b in (box_lig, box_bind, box_expr):
    draw_box(b)

# --- Titles (split the right title on two lines) ---
def title(txt, b, fs):
    x,y,w,h = b
    cx = x + w/2
    ty = y + h - 0.02
    plt.text(cx, ty, txt, fontsize=fs, fontweight="bold",
             ha="center", va="top", wrap=True)

title("Ligand Docking",                box_lig,  TITLE_FS_MAIN)
title("Androgen Receptor Binding",     box_bind, TITLE_FS_MAIN)
title("Gene Expression\n(ARKO vs WT)", box_expr, TITLE_FS_RIGHT)

# --- Body text ---
def body(txt, b, dy=0.12):
    x,y,w,h = b
    plt.text(x+0.02, y+h-dy, txt, fontsize=BODY_FS, ha="left", va="top", wrap=True)

body("â€¢ Testosterone\nâ€¢ DHT\nâ€¢ Ostarine", box_lig)
body("Strongest affinity:\nOstarine â‰ˆ â€“10 kcal/mol", box_bind)

body("â†“ Differentially expressed genes\n"
     "â†“ Pathway enrichment (Enrichr; KEGG 2021)",
     box_expr, dy=0.18)

body("Key Altered Pathways:\n"
     "â€¢ Folate biosynthesis\n"
     "â€¢ ABC transporters\n"
     "â€¢ Prolactin signaling\n"
     "â€¢ Cytokineâ€“cytokine interaction",
     box_expr, dy=0.40)

# --- Arrows (center-to-center, with extra gap) ---
arrow = dict(arrowstyle="->", lw=2, color="gray")

y_mid = box_lig[1] + box_lig[3]/2
plt.annotate("", xy=(box_bind[0]-0.01, y_mid),
             xytext=(box_lig[0]+box_lig[2]+0.01, y_mid),
             arrowprops=arrow)

y_mid2 = box_bind[1] + box_bind[3]/2
plt.annotate("", xy=(box_expr[0]-0.01, y_mid2),
             xytext=(box_bind[0]+box_bind[2]+0.01, y_mid2),
             arrowprops=arrow)

# tiny vertical arrow inside right box
plt.annotate("",
             xy=(box_expr[0]+0.02, box_expr[1]+box_expr[3]*0.60),
             xytext=(box_expr[0]+0.02, box_expr[1]+box_expr[3]*0.78),
             arrowprops=arrow)

# caption
plt.text(0.5, 0.09, "Graphical Abstract â€“ AR Knockout Muscle",
         fontsize=12, style="italic", ha="center")

out = "5_figures/graphical_abstract.png"
plt.tight_layout()
plt.savefig(out, dpi=300)
plt.show()
print("Saved:", out)
```

```python
import os, sys
from datetime import date

# Ensure the library is available
!{sys.executable} -m pip install -q python-pptx

from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor

# Paths
root = os.path.expanduser("~/Desktop/androgen_docking_project")
fig_dir = os.path.join(root, "5_figures")
out_pptx = os.path.join(fig_dir, "AR_Knockout_Summary.pptx")

fig1 = os.path.join(fig_dir, "graphical_abstract.png")
fig2 = os.path.join(fig_dir, "figure2_docking_x_pathways_color.png")
fig3 = os.path.join(fig_dir, "figure3_summary_plate.png")

for p in [fig1, fig2, fig3]:
    assert os.path.exists(p), f"Missing image: {p}"

# Create deck (16:9)
prs = Presentation()
prs.slide_width, prs.slide_height = Inches(13.33), Inches(7.5)

# ---- Slide 1: Title ----
title_slide = prs.slides.add_slide(prs.slide_layouts[5])  # blank, weâ€™ll add text
left, top, width, height = Inches(0.5), Inches(0.7), Inches(12.3), Inches(1.6)
title_box = title_slide.shapes.add_textbox(left, top, width, height)
tf = title_box.text_frame
tf.word_wrap = True
p = tf.paragraphs[0]
p.text = "AR Knockout Muscle â€” Docking & Transcriptomics"
p.font.size = Pt(40)
p.font.bold = True
p.alignment = PP_ALIGN.CENTER

sub = title_slide.shapes.add_textbox(Inches(2.0), Inches(2.1), Inches(9.3), Inches(0.9)).text_frame
sp = sub.paragraphs[0]
sp.text = "Independent Research Project\n(GSE216372 â€¢ Enrichr KEGG 2021 â€¢ Molecular Docking)"
sp.font.size = Pt(18)
sp.alignment = PP_ALIGN.CENTER

# Footer/date
footer = title_slide.shapes.add_textbox(Inches(0.5), Inches(6.9), Inches(12.3), Inches(0.5)).text_frame
fp = footer.paragraphs[0]
fp.text = f"Prepared {date.today().isoformat()}"
fp.font.size = Pt(12)
fp.font.color.rgb = RGBColor(120, 120, 120)
fp.alignment = PP_ALIGN.CENTER

# ---- Helper to add an image-centered slide with caption ----
def add_image_slide(image_path, caption):
    s = prs.slides.add_slide(prs.slide_layouts[6])  # blank
    # place image with margins
    margin_w, margin_h = Inches(0.5), Inches(0.6)
    avail_w = prs.slide_width - 2 * margin_w
    avail_h = prs.slide_height - 2 * margin_h - Inches(0.5)  # leave space for caption

    pic = s.shapes.add_picture(image_path, margin_w, margin_h, width=avail_w)
    # If image is too tall after width fit, scale it down to fit height
    if pic.height > avail_h:
        scale = avail_h / pic.height
        pic.height = int(pic.height * scale)
        pic.width  = int(pic.width * scale)
        # center horizontally
        pic.left = int((prs.slide_width - pic.width) / 2)

    # caption
    cap = s.shapes.add_textbox(Inches(0.5), prs.slide_height - Inches(0.65), prs.slide_width - Inches(1.0), Inches(0.5))
    t = cap.text_frame
    t.clear()
    para = t.paragraphs[0]
    para.text = caption
    para.font.size = Pt(14)
    para.alignment = PP_ALIGN.CENTER

# ---- Slide 2: Figure 1
add_image_slide(fig1, "Figure 1 â€” Graphical Abstract")

# ---- Slide 3: Figure 2
add_image_slide(fig2, "Figure 2 â€” Docking and Transcriptomic Pathway Integration")

# ---- Slide 4: Figure 3
add_image_slide(fig3, "Figure 3 â€” Summary Integration Plate")

# Save
prs.save(out_pptx)
print("Saved:", out_pptx)
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
import seaborn as sns  # seaborn only for the heatmap; fine to use here

# ---------- paths ----------
ROOT = Path.home() / "Desktop" / "androgen_docking_project"
GEX_DIR = ROOT / "3_gene_expression"
FIG_DIR = ROOT / "5_figures"
OUT_DIR = ROOT / "4_enrichment_outputs"

FIG_DIR.mkdir(parents=True, exist_ok=True)
OUT_DIR.mkdir(parents=True, exist_ok=True)

# find the excel file you downloaded
xlsx_files = list(GEX_DIR.glob("*.xlsx"))
assert len(xlsx_files) >= 1, f"No .xlsx found in {GEX_DIR}"
xlsx_path = xlsx_files[0]  # if more than one, pick the first; change if needed
print("Using:", xlsx_path.name)

# ---------- load ----------
df = pd.read_excel(xlsx_path)

# column sanity
needed = {"padj","log2FoldChange","Genename"}
missing = [c for c in needed if c not in df.columns]
assert not missing, f"Missing columns: {missing}"

# pick normalized expression columns for heatmap (WT_*_normalized & ARKO_*_normalized)
exp_cols = [c for c in df.columns if c.endswith("_normalized")]
assert len(exp_cols) >= 4, "Expected normalized expression columns not found."

# ---------- thresholds ----------
padj_thr = 0.05
l2fc_thr = 0.5

sig = df.dropna(subset=["padj"]).copy()
sig["sig"] = (sig["padj"] < padj_thr) & (sig["log2FoldChange"].abs() >= l2fc_thr)

up   = sig[(sig["sig"]) & (sig["log2FoldChange"] > 0)].copy()
down = sig[(sig["sig"]) & (sig["log2FoldChange"] < 0)].copy()

# keep useful columns
keep_cols = ["Genename","log2FoldChange","padj","baseMean"] + exp_cols
up_tidy   = up[keep_cols].sort_values("log2FoldChange", ascending=False)
down_tidy = down[keep_cols].sort_values("log2FoldChange", ascending=True)

# save lists
up_csv   = OUT_DIR / "deg_up_padj0.05_lfc0.5.csv"
down_csv = OUT_DIR / "deg_down_padj0.05_lfc0.5.csv"
up_txt   = OUT_DIR / "deg_up_for_enrichr.txt"
down_txt = OUT_DIR / "deg_down_for_enrichr.txt"

up_tidy.to_csv(up_csv, index=False)
down_tidy.to_csv(down_csv, index=False)
up_tidy["Genename"].dropna().to_csv(up_txt, index=False, header=False)
down_tidy["Genename"].dropna().to_csv(down_txt, index=False, header=False)

print("Saved:", up_csv.name, "|", down_csv.name)
print("Saved:", up_txt.name, "|", down_txt.name)

# ---------- volcano plot ----------
plt.figure(figsize=(8,6), dpi=180)
x = sig["log2FoldChange"]
y = -np.log10(sig["padj"].clip(lower=1e-300))  # avoid inf

# base points
plt.scatter(x, y, s=6, alpha=0.25, color="#9aa0a6", edgecolors="none", label="not significant")

# significant up/down
plt.scatter(up["log2FoldChange"], -np.log10(up["padj"]), s=12, alpha=0.8, color="#2ca02c", label="up")
plt.scatter(down["log2FoldChange"], -np.log10(down["padj"]), s=12, alpha=0.8, color="#d62728", label="down")

# thresholds lines
plt.axvline(l2fc_thr, color="#666", lw=1, ls="--")
plt.axvline(-l2fc_thr, color="#666", lw=1, ls="--")
plt.axhline(-np.log10(padj_thr), color="#666", lw=1, ls="--")

plt.xlabel("log2 Fold Change (ARKO vs WT)")
plt.ylabel("-log10(padj)")
plt.title("Volcano Plot â€” ARKO vs WT (muscle)")
plt.legend(frameon=False, loc="upper right")
plt.tight_layout()

volcano_path = FIG_DIR / "figure4_volcano.png"
plt.savefig(volcano_path)
plt.show()
print("Saved:", volcano_path)

# ---------- heatmap of top genes ----------
# select top 25 up and top 25 down (by padj then |lfc|)
top_up   = up_tidy.sort_values(["padj", "log2FoldChange"], ascending=[True, False]).head(25)
top_down = down_tidy.sort_values(["padj", "log2FoldChange"], ascending=[True, True]).head(25)
top_genes = pd.concat([top_up, top_down], axis=0)

hm = top_genes.set_index("Genename")[exp_cols]

# z-score per gene (row) for comparability
hm_z = (hm - hm.mean(axis=1).values[:,None]) / (hm.std(axis=1).replace(0, np.nan).values[:,None])
hm_z = hm_z.fillna(0)

plt.figure(figsize=(8,10), dpi=180)
sns.heatmap(hm_z, cmap="vlag", center=0, cbar_kws={"label":"Row z-score"})
plt.title("Top Differentially Expressed Genes â€” normalized expression (z-scored per gene)")
plt.xlabel("Samples")
plt.ylabel("Genes")
plt.tight_layout()

heatmap_path = FIG_DIR / "figure5_topgenes_heatmap.png"
plt.savefig(heatmap_path)
plt.show()
print("Saved:", heatmap_path)
```

```python
# Identify the gene-name column (usually "gene" or "Gene" or similar)
def pick_gene_col(df):
    for c in df.columns:
        if c.lower() in ["gene", "genes", "symbol", "gene_name", "genename"]:
            return c
    return df.columns[0]  # fallback

gene_col_up = pick_gene_col(deg_up)
gene_col_down = pick_gene_col(deg_down)

print("Gene column (up):", gene_col_up)
print("Gene column (down):", gene_col_down)

top_up = deg_up.sort_values("log2FoldChange", ascending=False).head(10)[[gene_col_up, "log2FoldChange", "padj"]]
top_down = deg_down.sort_values("log2FoldChange", ascending=True).head(10)[[gene_col_down, "log2FoldChange", "padj"]]

print("\nTop 10 UP:")
print(top_up.to_string(index=False))

print("\nTop 10 DOWN:")
print(top_down.to_string(index=False))
```
